## Imports and loading the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import warnings
import sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
train_df = pd.read_csv("../pythonProject1/Datasets/train.csv", nrows=1000000,
                    dtype={'acoustic_data': np.int16, 'time_to_failure': np.float64})

train_df.head()
```

## First look at the data

As we can see there are only two columns in the dataset: it's signal and earthquake time. By just loooking at the data we can clearly see that there are the same values for the quaketime in the first five examples

```python
train_df['time_to_failure'].value_counts().head(15)
```

```python
train_df['time_to_failure'].value_counts().tail(15)
```

Well, there are definetly different vLues. Lets see the distributions

```python
plt.figure(figsize=(10, 6))
sns.histplot(train_df['time_to_failure'], color='brown', kde=True, edgecolor='black', bins=int(180/5))
plt.xlabel('Quaketime')
plt.ylabel('Amount of values')
plt.title('Distribution of time_to_failure')
plt.show()
```
